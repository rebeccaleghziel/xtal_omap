In this notebook we input the orientation matches for the coccolith data acquired with 4D-STEM and solved using nbed, trackpy and py4dstem libraries.\
Here by using 2 classes (dataset and crystal group) together with a module to obtain stereographic projections, we obtain the full orientation maps, segmentation of single crystal units and access to the single crystal information for further data analysis

```python
import numpy as np
import matplotlib.pyplot as plt

from xtal_omap import Dataset
from xtal_omap.stereofunctions import *

import pandas as pd
import py4DSTEM
from emdfile import PointList
import math
```

## Defining the Dataset class
Dataset class uses the orientation matrix, correlation, pixel coordinates, bragg peaks, sweep size and crystal unit cell to:
1. Filter out the relevant pixels corresponding to the coccolith crystals and remove the backround
2. Cluster the pixels that have the same orientation within user specified bounderies
3. Plot and visualize the full coccolith map using stereogram

```python
folder_path1 = '/Users/houben/Tyche/git/xtal_omap/data'

#Inputs for the dataset class
orientation_matrix1 = np.load(f'{folder_path1}/orientation_matrix.npy')
corr_matrix1 = np.load(f'{folder_path1}/corr_matrix.npy')
pxl_coordinates1 = np.load(f'{folder_path1}/pxl_coordinates.npy')
bragg_peaks1 = np.load(f'{folder_path1}/braggpeaks_array.npy', allow_pickle=True)
sweep_size_in1 = [0, 255, 0, 216] # x1, x2, y1, y2
```

```python
cif_file_path = '/Users/houben/Tyche/git/xtal_omap/data/CalciteRhombohedral.cif'
calcite = py4DSTEM.process.diffraction.Crystal.from_CIF(cif_file_path)
k_max = 1.2
calcite.calculate_structure_factors(k_max)
calcite.setup_diffraction(200e3)
calcite.orientation_plan(
     angle_step_zone_axis = 1,
     angle_step_in_plane = 1,
     accel_voltage = 200e3,
     corr_kernel_size=0.08,
     zone_axis_range='auto',
    )
```

```python
# Initialize the dataset class

NBED_025 = Dataset.Dataset(
    crystal_class = calcite,
    sweep_size = sweep_size_in1,
    coordinates = pxl_coordinates1,
    orientation_matrix = orientation_matrix1,
    correlation = corr_matrix1,
    bragg_peaks = bragg_peaks1,
    )
```

## Filtering and Clustering
- Choose the correlation threshold manually by using plot_correlation method
- Set the thresholds for clustering: distance between pixels, the angle of the c-axis, and the minimum number of pixels to define a cluster
- Filter: "ON" if want to flip the mismatches of 180 degrees around x and y relevant for our system, "OFF" if cluster without flipping

```python
# Choose correlation threshold:
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 10))
NBED_025.plot_correlation(0.6, fig_handle = [fig, ax1])
NBED_025.plot_correlation(0.8, fig_handle = [fig, ax2])
NBED_025.plot_correlation(1, fig_handle = [fig, ax3])
```

```python
# Perform clustering 
# Manually choose the thresholds

NBED_025.run(corr_threshold = 0.8, dist_threshold = 30, theta_th = 3.5, min_groupsize = 50, filter = 'ON')
```

## Plotting and visualizing the overall coccolith architecture
Here we plot each cluster with a specific color in a user defined colormap and plot the stereographic projections of the a, b, c axis of all the clusters

```python
NBED_025.plot_map(colormap_type = 'Spectral', CoM = 'ON')
plt.savefig('high_resolution_figure_fullcocco.png', dpi=500)  # Adjust dpi as needed
```

```python
# Visualizing in 3D the a, b and c-axis of all the single crystals

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
origin = np.array([0, 0, 0])  # Origin for vectors

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

for i in range(0, len(NBED_025.crystal_unit)):
    
    vectorz = NBED_025.crystal_unit[i].Zlattice_avg
    vectorx = NBED_025.crystal_unit[i].Xlattice_avg
    vectory = NBED_025.crystal_unit[i].Ylattice_avg

    vector_c = scalevector(vectorz, 1) # returns a vector scaled to length but maintaining the same direction
    vector_a = scalevector(vectorx, 1) # returns a vector scaled to length but maintaining the same direction
    vector_b = scalevector(vectory, 1) # returns a vector scaled to length but maintaining the same direction

    ax.quiver(*origin, *vector_c, color='b', arrow_length_ratio=0.1)
    ax.quiver(*origin, *vector_a, color='r', arrow_length_ratio=0.1)
    ax.quiver(*origin, *vector_b, color='g', arrow_length_ratio=0.1)

ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])


ax.grid(False)

ax.view_init(elev=10, azim=30) 
```

## Accessing the data of single crystals

We use the methods of crystalgroup_v3 class to obtain plots and access relevant data of the single crystal units for further data analysis:
- Plotting of pixels corresponding to single crystals
- Accessing diffraction matches
- Histogram of correlation
- Stereographic projection and trace of the single clusters

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(25, 5))

NBED_025.crystal_unit[32].plot_single_crystal(ax = ax1, color_in = 'purple')
NBED_025.crystal_unit[33].plot_single_crystal(ax = ax1, color_in = 'green')

NBED_025.crystal_unit[32].plot_stereo(axis = ax2)
NBED_025.crystal_unit[32].add_trace(ax = ax2, color = 'purple')

NBED_025.crystal_unit[33].plot_stereo(axis = ax2)
NBED_025.crystal_unit[33].add_trace(ax = ax2, color = 'green')

NBED_025.crystal_unit[32].diffraction_match(fig = fig, ax = ax3, n_diff = 20)
NBED_025.crystal_unit[33].diffraction_match(fig = fig, ax = ax4, n_diff = 20)

ax3.set_title("Cluster 32")
ax4.set_title("Cluster 33")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3))

NBED_025.crystal_unit[32].histogram_of_correlation(ax = ax1)
NBED_025.crystal_unit[33].histogram_of_correlation(ax = ax2)
```
